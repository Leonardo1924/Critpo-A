**Q2: Shifting the alphabet**

Consider the following encryption scheme(E, D), with message and ciphertext space the english alphabet, considering words of size n. 
The scheme is as follows:

- Generate a key k with n uniform values[0. . .25]
- E(k, m)shifts the letters of m according to k, producing c
- D(k, m) takes c and applies the reverse shift according to k

A simple example of how this encryption scheme works:
- k={3,7,1,20,15,2}
- m=banana
- c=ehoucc

Question: Is the proposed scheme E perfectly secure?

> To assess whether the proposed encryption scheme is perfectly secure, we have to consider the perfect secure definition established by Claude Shannon.
>
>DEF: A encryption scheme is considered perfectly secure if, for each message M and each Ciphertex C, the probability of M data C remains constant, independently of the value of C, that is:
>
> $$ ∀m,c. P(M = m | C = c) = P (M = m)$$
>
> Returning to the proposed scheme, there is a K Key that consists of n evenly chosen values ​​of the $[0,...,25]$ corresponding to the 26 letters of the alphabet. Operation E (K, m) shifts the letters of message m to the corresponding values ​​in key k. Consequently, the function D (k, c) applies a reverse displacement based on the same key
>
> 1. Key Gen -> The key K has n values, each one uniformly chosen, resultin in $26^N$ possible keys.
>
> 2. M and C -> This to texts belong to the same alfabetic space of K, so there are $26^N$ possible combinations.
>
>
> Each text character is displaced independently by the corresponding value of the key. This means that for any character given a text there are 26 possible displacements, resulting in a uniform distribution of ciphertexts to a particular plaintext. The crucial factor for a Scheme perfect secure lies in the ratio between the number of plaintexts and the number of Ciphertexts.
>
> Since the keys are randomly uniform, but the plaintext can overlap in their displaced representations, we can find instances in which multiple plaintexts generate the same Ciphertext, with such collisions we cannot say that the scheme is perfectly secure.
